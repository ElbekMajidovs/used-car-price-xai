# 04 — Explainable AI (XAI) Analysis
## Used Car Price Prediction

**Purpose:** Interpret the best-performing model using multiple XAI techniques to understand *why* the model makes specific price predictions.

### XAI Techniques Covered

| Method | Scope | Library |
|---|---|---|
| SHAP Summary Plot | Global | `shap` |
| SHAP Waterfall / Force Plot | Local (per prediction) | `shap` |
| SHAP Interaction Values | Pairwise feature interactions | `shap` |
| LIME | Local | `lime` |
| Partial Dependence Plot (PDP) | Global — marginal effect | `sklearn` |
| Individual Conditional Expectation (ICE) | Instance-level effect | `sklearn` |
| Permutation Feature Importance | Global | `sklearn` |

### Business Narrative
For each XAI output we will translate the technical finding into a plain-language insight:
- e.g. *"Each additional year of age reduces a car's expected price by roughly 8%, all else equal."*
- e.g. *"This specific listing is priced $2,400 above model prediction primarily because it has only 23K miles."*
